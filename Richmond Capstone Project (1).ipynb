Breaking down each Zip Code by venues

```python
RVA_onehot = pd.get_dummies(dfk[['Venue']], prefix="", prefix_sep="")

RVA_onehot['Zip Code'] = dfk['Zip Code']

fixed_columns = [RVA_onehot.columns[-1]] + list(RVA_onehot.columns[:-1])
RVA_onehot = RVA_onehot[fixed_columns]

RVA_onehot.head()
```

```python
RVA_onehot.shape
```

```python
dfkm = RVA_onehot.groupby('Zip Code').mean().reset_index()
dfkm
```

```python
dfkm.shape
```

```python
def return_most_common_venues(row, top_two):
    row_categories = row.iloc[1:]
    row_categories_sorted = row_categories.sort_values(ascending=False)
    
    return row_categories_sorted.index.values[0:top_two]
```

Top Venue for each Zip Code

```python
top_two = 2

indicators = ['st', 'nd']

columns = ['Zip Code']
for ind in np.arange(top_two):
    try:
        columns.append('{} Most Common Venue'.format(ind+1, indicators[ind]))
    except:
        columns.append('{} Most Common Venue'.format(ind+1))
        
dfks = pd.DataFrame(columns=columns)
dfks['Zip Code'] = dfkm['Zip Code']

for ind in np.arange(dfkm.shape[0]):
    dfks.iloc[ind, 1:] = return_most_common_venues(dfkm.iloc[ind, :], top_two)
    
dfks.head()
```

```python
kclusters = 4

dfkms = dfkm.drop('Zip Code', 1)

kmeans = KMeans(n_clusters = kclusters, random_state=0).fit(dfkms)

kmeans.labels_[0:2]
```
